### <font color="midnightblue"> Data Cleaning 1 </font>

```python
# Subset to df1 which will contain the data for the holidays that will be comparing 

df[(df['ARREST_DATE'] == '01/01/2024') | 
   (df['ARREST_DATE'] == '02/13/2024') | 
   (df['ARREST_DATE'] == '02/14/2024') | 
   (df['ARREST_DATE'] == '03/17/2024')]

df1 = df[(df['ARREST_DATE'] == '01/01/2024') | 
   (df['ARREST_DATE'] == '02/13/2024') | 
   (df['ARREST_DATE'] == '02/14/2024') | 
   (df['ARREST_DATE'] == '03/17/2024')]

df1
```

```python
# Subset to df2 which will contain only the columns interested in viewing 

df1 [['ARREST_KEY', 'ARREST_DATE', 'OFNS_DESC', 'ARREST_BORO', 'AGE_GROUP', 'PERP_SEX', 'PERP_RACE']]
df2 = df1 [['ARREST_KEY', 'ARREST_DATE', 'OFNS_DESC', 'ARREST_BORO', 'AGE_GROUP', 'PERP_SEX', 'PERP_RACE']]

df2
```

```python
# Renaming Columns

df2 = df2.rename(columns={'ARREST_KEY' : 'ARREST #', 'ARREST_DATE' : 'ARREST DATE', 'OFNS_DESC' : 'OFFENSE', 'AGE_GROUP' : 'AGE GROUP', 'ARREST_BORO' : 'BOROUGH', 'PERP_SEX' : 'GENDER', 'PERP_RACE' : 'RACE'})

df2.head(2)
```

### <font color="midnightblue"> Visualization 1 </font>

```python
# Countplot allows visualization the # of crimes commited each holiday

sns.countplot (x = 'ARREST DATE', data = df2)
```

```python
# Added hue to add a layer of data for gender, assigned color palette, changed background to dark grid

sns.countplot (x = 'ARREST DATE', hue = 'GENDER', palette = "bone", data = df2)
sns.set_style('darkgrid')
```

```python
# Added bar count labels, increased font, added title, renamed x + y labels 

ax = sns.countplot (x = 'ARREST DATE', hue = 'GENDER', palette = "bone", data = df2)
for container in ax.containers:
    ax.bar_label(container)
sns.set_context('notebook', font_scale = 1)
ax.set (title = 'Arrests based on Holiday and gender', xlabel = 'Arrest Date', ylabel = 'Arrest Count')
```

### <font color="midnightblue"> NOTE: </font>

<p> <strong> <font color="midnightblue"> Based on the visualization above, the holiday with the highest count of arrests is Valentines Day, with the majority of the arrests being for the Male population. Due to the disparity of arrests for one holiday, further research was performed on the top 3 types of offenses for each holiday. </font> </strong>

### <font color="midnightblue"> Data Exploration II - Detailed </font> 

```python
# Subsetting to df3 to get output for the type of offenses on 02/14/2024 Valentines day

df2[df2['ARREST DATE'] == '02/14/2024']
df3 = df2[df2['ARREST DATE'] == '02/14/2024']

df3
```

```python
# Subsetting to df4 to get top 3 offenses on 02/14/2024 Valentines day

df3.groupby (['OFFENSE'])['ARREST DATE'].count().sort_values(ascending=False).head(3)
df4 = df3.groupby (['OFFENSE'])['ARREST DATE'].count().sort_values(ascending=False).head(3)

df4
```

```python
# Subsetting to df5 to get output for the type of offenses on 02/13/2024 Mardi Gras

df2[df2['ARREST DATE'] == '02/13/2024']
df5 = df2[df2['ARREST DATE'] == '02/13/2024']

df5
```

```python
# Subsetting to df6 to get top 3 offenses on 02/13/2024 Mardi Gras

df5.groupby (['OFFENSE'])['ARREST DATE'].count().sort_values(ascending=False).head(3)
df6 = df5.groupby (['OFFENSE'])['ARREST DATE'].count().sort_values(ascending=False).head(3)

df6
```

```python
# Subsetting to df7 to get output for the type of offenses on 03/17/2024 St. Patricks Day

df2[df2['ARREST DATE'] == '03/17/2024']
df7 = df2[df2['ARREST DATE'] == '03/17/2024']

df7
```

```python
# Subsetting to df8 to get top 3 offenses on 03/17/2024 St. Patricks Day

df7.groupby (['OFFENSE'])['ARREST DATE'].count().sort_values(ascending=False).head(3)
df8 = df7.groupby (['OFFENSE'])['ARREST DATE'].count().sort_values(ascending=False).head(3)

df8
```

```python
# Subsetting to df9 to get output for the type of offenses on 01/01/2024 New Years Day

df2[df2['ARREST DATE'] == '01/01/2024']
df9 = df2[df2['ARREST DATE'] == '01/01/2024']

df9
```

```python
# Subsetting to df10 to get top 3 offenses on 01/01/2024 New Years Day

df9.groupby (['OFFENSE'])['ARREST DATE'].count().sort_values(ascending=False).head(3)
df10 = df9.groupby (['OFFENSE'])['ARREST DATE'].count().sort_values(ascending=False).head(3)

df10
```

### <font color="midnightblue"> Data Cleaning 2 </font>

```python
# Joined/Concat data frames with keys labeled as the holiday

(pd.concat([df4, df6, df8, df10], keys=['02/14/2024', '02/13/2024', '03/17/2024', '01/01/2024']))
```

```python
# Subset to df11 to its own data frame

df11 = (pd.concat([df4, df6, df8, df10], keys=['02/14/2024', '02/13/2024', '03/17/2024', '01/01/2024']))

df11
```

### <font color="midnightblue"> Visualization 2 </font>

```python
# Barh plot allows visualization of the # of top 3 arrests commited each holiday

df11.plot(kind='barh')
plt.title("TOP OFFENSES/ARRESTS PER HOLIDAY")
plt.xlabel("ARRESTS")
plt.ylabel("DATE & OFFENSE")
```
