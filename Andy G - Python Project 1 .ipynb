```python
# Data Cleaning df - Reloading DF with Non Standard Missing Values "s" & "na" to Standard NaN, and dropping NaN values


missing_values = ["s","na"]
df = pd.read_csv ('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv',
                na_values = missing_values).dropna()

df
```

```python
# Data Cleaning df - Changing data type from object to int to columns will be focusing on  


df[['Number Scoring Below 65','Number Scoring 65 or Above','Number Scoring 80 or Above','Number Scoring CR']] = df[['Number Scoring Below 65','Number Scoring 65 or Above','Number Scoring 80 or Above','Number Scoring CR']].astype(int)

df
```

```python
df.dtypes
```

```python
# Data Cleaning df - Subsetting to df2 for output with only desired columns in output


df[['School DBN','School Name','School Level','Regents Exam','Year','Total Tested','Mean Score','Number Scoring Below 65','Number Scoring 65 or Above','Number Scoring 80 or Above','Number Scoring CR']]
df2 = df[['School DBN','School Name','School Level','Regents Exam','Year','Total Tested','Mean Score','Number Scoring Below 65','Number Scoring 65 or Above','Number Scoring 80 or Above','Number Scoring CR']]

df2
```

```python
# Data Exploration & Cleaning df2 - Subsetting to df3 for output with only one school level to be compared (High School)


df2['School Level'].unique()
```

```python
df2[df2['School Level'] == 'High school']
df3 = df2[df2['School Level'] == 'High school']

df3
```

```python
# Data Exploration & Cleaning df3 - Subsetting to df4 for output with only one year to be compared (2015)


df3['Year'].unique()
```

```python
df3[df3['Year'] == 2015]
df4 = df3[df3['Year'] == 2015]

df4
```

```python
# Data Exploration & Cleaning df4 - Subsetting to df5 for output with one test to be compared (Common Core Algebra)


df4['Regents Exam'].unique()
```

```python
df4[df4['Regents Exam'] == 'Common Core Algebra']
df5= df4[df4['Regents Exam'] == 'Common Core Algebra']

df5
```

```python
# Data Exploration & cleaning df5 - Filtered "Total Tested" values over 500 and Subsetting to df6 for selected schools to compare (Harry S Truman 75710 + Fort Hamilton High School 11203)
# Schools picked due to closest # of "Total Tested" : 7510 @ 704 + 12203 @ 707


df5[df5['Total Tested'] >= 500].head()
```

```python
df6 = df.filter(items=[7510, 12203], axis=0)

df6
```

```python
# Data Cleaning df6 - Subsetting to df7 for Visualization output 


df6[['School Name','Total Tested','Number Scoring Below 65','Number Scoring 65 or Above','Number Scoring 80 or Above']]
df7 = df6[['School Name','Total Tested','Number Scoring Below 65','Number Scoring 65 or Above','Number Scoring 80 or Above']]

df7
```

```python
# Data Visualization df7 - Bar


df7.plot(kind = "bar")
```

```python
# Data Visualization df7 - Barh


df7.plot(kind = "barh", title = 'Comparison of Common Core Algebra Exam Scores', x = 'School Name')
```
